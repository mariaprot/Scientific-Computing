```python
# Define the function and its derivative
def f(x):
    return x * np.sin(3 * x) - np.exp(x)

def df(x):
    return np.sin(3 * x) + 3 * x * np.cos(3 * x) - np.exp(x)

# Set up the plot
x_values = np.linspace(-1.5, 0, 400)
y_values = f(x_values)

# Part 1: Newton-Raphson Method
x_n = -1.6  # initial guess
tolerance = 1e-6
max_iterations = 100
iterations_n = 0

for j in range(max_iterations):
    f_n = f(x_n)
    if abs(f_n) < tolerance:
        newton_root = x_n
        break
    df_n = df(x_n)
    x_n = x_n - f_n / df_n  # update step
    iterations_n += 1
else:
    newton_root = None

print('Newton method:',newton_root)
print(iterations_n)

# Part 2: Bisection Method
xl = -0.7  # left endpoint
xr = -0.4  # right endpoint
iterations_b = 0

for iterations_b in range(max_iterations):
    xc = (xr + xl) / 2  # midpoint
    f_c = f(xc)
    
    if f_c == 0:  # exact solution found
        bisection_root = xc
        break
    elif f(xl) * f_c < 0:  # root is in left subinterval
        xr = xc
    else:  # root is in right subinterval
        xl = xc

    # Check for convergence
    if (xr - xl) / 2 < tolerance:
        bisection_root = (xr + xl) / 2
        break
else:
    bisection_root = None
print('Bisection method:', bisection_root)
print('iterations:',iterations_b)
```

```python
# Plotting the function
plt.figure(figsize=(10, 5))
plt.plot(x_values, y_values, label='f(x) = x sin(3x) - exp(x)', color='blue')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # Add x-axis
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')  # Add y-axis

# Highlight the roots found by both methods
if newton_root is not None:
    plt.plot(newton_root, f(newton_root), 'ro', label='Newton-Raphson root')
if bisection_root is not None:
    plt.plot(bisection_root, f(bisection_root), 'go', label='Bisection root')

plt.title('Plot of f(x) with Roots Found by Newton-Raphson and Bisection Methods')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.ylim(-5, 5)  # Adjust y-limits for better visualization
plt.show()
```
